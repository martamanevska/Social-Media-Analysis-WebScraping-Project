```python
# Import the two main libraries
import requests
import urllib.request,sys,time
import pandas as pd
from tqdm import tqdm
import pickle
import datetime
import json
```

The way that we scraped from the APIs is the same as [Reuters notebook](./scraper_reuters.ipynb), look at that one for more informations.

```python
#get all the links of the articles

saved_links = []
numberOfArticles = 0

# scraping data directly from the api's of CNN
for page in tqdm(range(0,91)):
    url = 'https://search.api.cnn.com/content?q=russia-ukraine&size=50&from='+str(numberOfArticles)+'&page='+str(page)+'&sort=newest'
    result = requests.get(url).text
    numberOfArticles += 50
    
    saved_links.append(result)
```

```python
file_name = "cnn_links.pkl"

open_file = open(file_name, "wb")
pickle.dump(saved_links, open_file)
open_file.close()
```

```python
# to load
open_file = open("cnn_links.pkl", "rb")
saved_links = pickle.load(open_file)
open_file.close()
```

```python
#creating the dataframe

CNNArticles = pd.DataFrame({'title': pd.Series([], dtype='string'),
                             'author': pd.Series([], dtype='string'),
                             'date': pd.Series([], dtype='float'),
                             'text': pd.Series([], dtype='string'),
                             })
```

Luckily CNN sends all the data that we need on those APIs so we only have to read what's inside the JSON and put that data on the dataset.

```python
for i, link in tqdm(enumerate(saved_links)):
    result = json.loads(link) # Since the api's gives back a json file we need to load as a json in order to access to the objects inside
    
    for j in range(0,50): # we have a object called result, with everything we are looking for inside
        date = pd.to_datetime(result['result'][j]['firstPublishDate']).strftime('%d/%m/%y') # here we have the publish date
        author = result['result'][j]['contributors'] # here the name of alla the journalist
        title = result['result'][j]['headline'] # here the title
        text = result['result'][j]['body'] # here the body of each articles
        
        CNNArticles = CNNArticles.append({"title": title, "author": author, "date": date, "text": text}, ignore_index=True)
```
